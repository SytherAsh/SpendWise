```python
from groq import Groq
from tqdm import tqdm
import pandas as pd

# Initialize tqdm
tqdm.pandas()

# Set your Groq API key
groq_api = "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"
client = Groq(api_key=groq_api)

# Function to generate a note based on the description
def assign_note_with_groq(description):
    prompt = f"""
    Categorize the following transaction description into one of the following tags: 
    ["Grocery", "Entertainment", "Shopping", "Food", "Utilities", "Rent", "Travel", "Others"]. 
    If the category is unclear, assign "Others". 
    Description: "{description}" 
    Category:
    """
    try:
        # Call Groq's LLM API
        chat_completion = client.chat.completions.create(
            messages=[
                {
                    "role": "user",
                    "content": prompt,
                }
            ],
            model="gemma2-9b-it",  # Specify the model
        )
        # Extract the category from the response
        category = chat_completion.choices[0].message.content.strip()
        return category
    except Exception as e:
        print(f"Error: {e}")
        return "Error"

# Apply the function with tqdm progress bar
df["Note"] = df["Description"].progress_apply(assign_note_with_groq)

# Display the resulting DataFrame
print(df)
```
