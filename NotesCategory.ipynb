```python
import pandas as pd
from tqdm import tqdm

def map_to_company(text):
    for company, keywords in company_keywords.items():
        # Perform a case-insensitive check for each keyword
        if any(keyword.lower() in text for keyword in keywords):
            return company
    # Return None if no mapping is found
    return None

# Create merged_text by combining Recipient_Name and UPI_ID
df['merged_text'] = (
    df['Recipient_Name'].fillna('').str.strip().str.lower() + " " + 
    df['UPI_ID'].fillna('').str.strip().str.lower()
).str.strip()

tqdm.pandas(desc="Mapping Companies")
df['Company'] = df['merged_text'].progress_apply(map_to_company)

# Filter to keep only rows where Company is not None (i.e. a mapping was found)
df_categorized = df[df['Company'].notnull()].copy()

print("✅ Processing complete! Only categorized companies retained.")
```
